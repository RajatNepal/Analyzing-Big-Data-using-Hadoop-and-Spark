```python
%%bash

#creating the directory if it does not already exist
if [ ! -d "wikipedia-ml-raw" ]; 
  then mkdir wikipedia-ml-raw; 
fi

#opening given txt file and going line by line
cat article-ids.txt | while read line; 
do
  #gparsing the articleId from the line and the name of file it will be saved as
  articleId=$(echo $line | tr -d '\r')
  filename="machine-learning-$articleId.html"
  url="https://en.wikipedia.org/w/index.php?title=Machine_learning&oldid=$articleId"
#this downloads the files (doesnt download the file without articleID since there is white space in the txt file)
  if [ $filename != "machine-learning-.html" ]; 
    then wget -O "wikipedia-ml-raw/$filename" "$url"; 
  fi
done
```
